```python
from ultralytics import YOLO
import os
```

Primeiro vamos testar o modelo padr√£o do Yolo nas nossasa imgens de teste

```python
coco_animal_classes = {
    14: 'bird',
    15: 'cat',
    16: 'dog',
    17: 'horse',
    18: 'sheep',
    19: 'cow',
    20: 'elephant',
    21: 'bear',
    22: 'zebra',
    23: 'giraffe'   
}
```

```python

model = YOLO('yolov8m.pt')
dir_path = './test_images'
folders = [f for f in os.listdir(dir_path) if os.path.isdir(f'{dir_path}/{f}')]

for folder in folders:
    results = model.predict(f'./test_images/{folder}'
                            , classes = list(coco_animal_classes.keys())
                            , save=True
                            , save_txt=True
                            , save_conf=True
                            , conf = 0.5
                            , max_det = 5)
```

```python
dir_path = './runs/detect/predict/labels'
files = [f for f in os.listdir(dir_path) if os.path.isfile(f'{dir_path}/{f}')]

for key, animal in coco_animal_classes.items():
    confidence_sum = 0
    total_animals = 0
    for file in files:
        if file.startswith(animal):
            with open(f'{dir_path}/{file}', 'r') as f:
                print(f'{file}:')
                for line in f.readlines():
                    line = line.split(' ')
                    if line[0] == str(key):
                        print(f'    classe {line[0]}-{coco_animal_classes[key]} -> confidencia {line[-1][:-1]}')
                        confidence_sum += float(line[-1])
                        total_animals += 1
    confidence = confidence_sum / total_animals

    print('-'*40)
    print(f'for {total_animals} {animal}: {confidence:.2f}')
    print('-'*40)
    print()          
```
